## Dataframes de Pandas

### Introducción

Un DataFrame de Python es una tabla 2-dimensional, potencialmente hetereogéna y con ejes etiquetados (filas y columnas).

Los DataFrames de Pandas consisten básicamente de 3 elementos principales:
-  datos (*data*)
-  filas (*rows*)
-  columnas (*columns*)

![title](https://media.geeksforgeeks.org/wp-content/uploads/finallpandas.png)

**Importante!** La notación nos recomienda siempre que creemos DataFrames darles nombres con sentido, derivados del acrónimo **d**ata**f**rame: df, df1, df2 etc

|Parámetro|Descripción|
|--------|------------|
|**data**| Los datos que incorpora como unpit pueden tener varias formas como: *ndarray*, *series*, *list*, *dict*, o otros dataframes
|**index**| Al ser una estructura 2-d tanto las filas como las columnas llevan etiquetas, siendo **las etiquetas de las filas** predeterminadas a *np.arange(n)* si no se especifican de otra forma|
|**columns**| Las etiquetas de las columnas están prestablecidas a *np.arange(n)* si no se especifican de otra forma|
|**dtype**| Tipo de datos de **cada** columan|
|**copy**| Este comando se utiliza para copiar los datos inputados. Por default está a False

**Tipos de datos en un Dataframe (dtypes)**

En un dataframe de pandas, el atributo dtype (abreviatura de "data type") se utiliza para especificar el tipo de datos de cada columna. Hay varios valores posibles para el dtype, que corresponden a los diferentes tipos de datos admitidos por pandas. Algunos de los valores posibles para dtype en un dataframe de pandas incluyen:

- int64: Tipo entero de 64 bits.
- float64: Tipo de punto flotante de precisión doble.
- object: Tipo genérico, generalmente utilizado cuando los datos en una columna son de tipos mixtos o no se pueden categorizar fácilmente en otros tipos de datos.
- bool: Tipo booleano, representando True (Verdadero) o False (Falso).
- datetime64: Tipo para fechas y horas.
- timedelta: Tipo utilizado para representar diferencias de tiempo.
- category: Tipo para datos categóricos, útil cuando hay un número finito y conocido de valores posibles.



Algunos de los tipos de datos para enteros y flotantes :

**Para enteros (int)**:

- int8: Tipo entero de 8 bits. Rango de valores de -128 a 127.
- int16: Tipo entero de 16 bits. Rango de valores de -32,768 a 32,767.
- int32: Tipo entero de 32 bits. Rango de valores de -2,147,483,648 a 2,147,483,647.
- int64: Tipo entero de 64 bits. Rango de valores de -9,223,372,036,854,775,808 a 9,223,372,036,854,775,807.

**Para números de punto flotante (float)**: float16, float32, float64, float128

#### Atributos

| Atributo | Descripción                                                                                             |
|----------|---------------------------------------------------------------------------------------------------------|
| shape    | Retorna una tupla que representa la forma del DataFrame (número de filas, número de columnas).         |
| dtypes   | Retorna un objeto que describe los tipos de datos de cada columna en el DataFrame.                      |
| index    | Retorna el índice de las filas del DataFrame.                                                          |
| columns  | Retorna las etiquetas de las columnas del DataFrame.                                                   |
| values   | Retorna una representación NumPy de los datos contenidos en el DataFrame.                               |
| size     | Retorna el número total de elementos en el DataFrame (número de filas * número de columnas).            |
| empty    | Retorna True si el DataFrame está vacío (sin filas ni columnas), False en caso contrario.              |
| ndim     | Retorna el número de dimensiones de los datos, siempre 2 para DataFrames.                               |
| axes     | Retorna una lista que contiene las etiquetas de los ejes del DataFrame (índice y columnas).             |
| T        | Retorna una transposición del DataFrame, es decir, las filas se convierten en columnas y viceversa.    |


#### Métodos

| Método        | Descripción                                                                                              |
|---------------|----------------------------------------------------------------------------------------------------------|
| head()        | Retorna las primeras filas del DataFrame.                                                                |
| tail()        | Retorna las últimas filas del DataFrame.                                                                  |
| info()        | Proporciona una visión general de la estructura y la información de los datos en el DataFrame.          |
| describe()    | Calcula estadísticas descriptivas para todas las columnas numéricas en el DataFrame.                      |
| mean()        | Calcula la media de las columnas numéricas en el DataFrame.                                               |
| median()      | Calcula la mediana de las columnas numéricas en el DataFrame.                                             |
| std()         | Calcula la desviación estándar de las columnas numéricas en el DataFrame.                                 |
| min()         | Encuentra el valor mínimo para cada columna numérica en el DataFrame.                                     |
| max()         | Encuentra el valor máximo para cada columna numérica en el DataFrame.                                     |
| sum()         | Calcula la suma de los valores para cada columna numérica en el DataFrame.                                 |
| isnull()      | Devuelve un DataFrame booleano indicando valores nulos.                                                    |
| dropna()      | Elimina filas o columnas con valores nulos.                                                                |
| fillna()      | Rellena los valores nulos con un valor específico.                                                         |
| astype()      | Cambia el tipo de datos de las columnas en el DataFrame.                                                   |
| sort_values() | Ordena el DataFrame según los valores de una o más columnas.                                               |
| groupby()     | Agrupa los datos según una o más columnas.

#### LOC y ILOC

Son muy útiles para acceder y manipular datos en un DataFrame. Su elección  depende de si se desea acceder a los datos por etiquetas o por índices enteros, respectivamente.

- **loc**: Este método se utiliza para acceder a un grupo de filas y columnas por etiquetas o una matriz booleana. Se utiliza principalmente para la selección basada en etiquetas.
- **iloc**: Este método se utiliza para acceder a un grupo de filas y columnas por índices enteros. Se utiliza principalmente para la selección basada en índices enteros.
