30 - Crea un diccionario llamado 'precios'.

- Añade los siguientes elementos al diccionario:
"plátano": 4,
"manzana": 2,
"naranja": 1.5,
"pera": 3

- Crea un diccionario llamado stock con la siguiente información:
"plátano": 6,
"manzana": 0,
"naranja": 32,
"pera": 12
- Calcula el dinero que ganarías si vendieses todo tu stock al precio marcado


```python
precios = {}
```

```python
precios["plátano"] = 4
precios["manzana"] = 2
precios["naranja"] = 1.5
precios["pera"] = 3
precios
```

```python
stock = {"plátano": 6, "manzana": 0, "naranja": 32, "pera": 12}
stock
```

```python
fruta = 'pera'
print(f"fruta: {fruta} \nprecio: {stock[fruta]} \nstock: {precios[fruta]}")
```

```python
sum((stock['plátano']*precios['plátano'],
     stock['pera']*precios['pera'],
     stock['manzana']*precios['manzana'],
     stock['naranja']*precios['naranja']))
```
